```python
import pandas as pd
import matplotlib.pyplot as plt

# -----------------------
# 1️⃣ Larger Sample Data
# -----------------------
data = {
    'Category': [
        'Food', 'Transport', 'Shopping', 'Bills', 'Entertainment',
        'Medicine', 'Education', 'Travel', 'Recharge', 'Others'
    ],
    'Amount': [4500, 1500, 3200, 2800, 1200, 900, 3500, 5000, 400, 600]
}

df = pd.DataFrame(data)
print("---- DATA ----")
print(df)

# -----------------------
# 2️⃣ Summary Statistics
# -----------------------
print("\n---- SUMMARY ----")
print(df.describe())

# -----------------------
# 3️⃣ Pie Chart
# -----------------------
plt.figure(figsize=(7,7))
plt.pie(
    df['Amount'],
    labels=df['Category'],
    autopct='%1.1f%%',
    startangle=140
)
plt.title("Expense Distribution - Pie Chart")
plt.show()

# -----------------------
# 4️⃣ Bar Chart
# -----------------------
plt.figure(figsize=(10,5))
plt.bar(df['Category'], df['Amount'])
plt.title("Expense Comparison - Bar Chart")
plt.xlabel("Category")
plt.ylabel("Amount")
plt.xticks(rotation=45)
plt.show()

# -----------------------
# 5️⃣ Line Chart
# -----------------------
plt.figure(figsize=(10,5))
plt.plot(df['Category'], df['Amount'], marker='o')
plt.title("Expense Trend - Line Chart")
plt.xlabel("Category")
plt.ylabel("Amount")
plt.grid(True)
plt.xticks(rotation=45)
plt.show()
```
